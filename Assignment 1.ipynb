```python
import math 

a = 2
b = 8
c = 1

d = (b * b) - (4 * a * c)
```

```python
x1 = (-b + math.sqrt(d))/ (2 * a)
print("x1 is {:.2f}".format(x1))

x2 = (-b - math.sqrt(d))/ (2 * a)
print("x1 is {:.2f}".format(x2))
```
